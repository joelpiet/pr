```python
# having range of numbers between 2 numbers (start, end, no range)
b=np.linspace(1,2,num=5)
print(b)
#if you dont want end number ie 2
c=np.linspace(1,2,num=5,endpoint=False)
print(c)
```

```python
d=np.array([np.arange(1,5),np.arange(11,15)])
print(d)
print(d[0,2])
```

```python
# selecting the subsets from the above 2D array using slicing
e=np.array(d[:,1:3])
print(e)
```

```python
f=np.array([np.arange(1,37)])
print(f)
```

```python
# reshaping the array to 3D
g=f.reshape(3,4,3) #np.reshape(f,(3,4,3))
print(g)
```
